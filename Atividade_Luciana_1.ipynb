 7- Traduzir os nomes das colunas para português Brasil.

```python
df.columns = [
    'id_do_programa', 'tipo', 'titulo', 'diretor', 'elenco', 'pais',
    'data_adicao', 'ano_lancamento', 'classificacao', 'duracao', 'listado_em', 'descricao'
]
print("Colunas renomeadas com sucesso para:")
print(df.columns)
```

9- Identificar a quantidade de Filmes e Séries na base.

```python
df['tipo'].value_counts()
```

10- Mostrar os 10 países mais frequentes na amostra.

```python
df['pais'].value_counts(10)
```

11- Verificar qual é a classificação indicativa mais comum.

```python
df['classificacao'].value_counts()
```

 12- Listar os 10 gêneros mais recorrentes.

```python
df['listado_em'].value_counts()
```

 13- Resumir as informações das variáveis categóricas.

```python
print("Resumo das Variáveis Categóricas:\n")

for col in df.select_dtypes(include='object').columns:
    print(f"Variável: {col}")
    print(f"  - Número de valores únicos: {df[col].nunique()}")
    print(f"  - Valor mais frequente: {df[col].mode()[0] if not df[col].mode().empty else 'N/A'}")
    print(f"  - Contagem do valor mais frequente: {df[col].value_counts().max() if not df[col].value_counts().empty else 'N/A'}")
    print(f"  - Percentual de valores ausentes: {df[col].isnull().sum() / len(df) * 100:.2f}%\n")
```
